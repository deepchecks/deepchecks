```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from deepchecks import Dataset

df = load_breast_cancer(as_frame=True).frame
label = 'target'

train_df, test_df = train_test_split(df, test_size=0.33)
train = Dataset(train_df, label=label)
test = Dataset(test_df, label=label)

clf = AdaBoostClassifier()
features = train_df.drop(label, axis=1)
target = train_df[label]
clf.fit(features, target)
```

```python
from deepchecks.checks import ConfidenceChange

ConfidenceChange(min_test_samples=100, bin_size=0.05).run(train, test, clf)
```
